```python
import yfinance as yf
import pandas as pd
from ipywidgets import Output
```

```python
stocks = ["aapl","avgo","msft"]
data =pd.DataFrame([])
pd.options.display.float_format = '{:.2f}'.format
pd.set_option('display.max_rows', None)
               
for stock in stocks:
    stk = yf.Ticker(stock)  
    stock_data = stk.history(period='1y')
    stock_data.insert(0,"Symbol",stock )
    # stock_data.loc[stock_data.index[-1],"End"]=True  
    data = pd.concat([data,stock_data],ignore_index = True)
del data["Dividends"]
del data["Stock Splits"] 
data["EMA"] = data["Volume"].ewm(span=20, adjust=False).mean().astype(int)

data["Target_10"] = data["Close"].shift(-5)
data["Entry"]= 0

```

```python
data["Entry"] =  ((data["Volume"]>data["EMA"])&
                 (data["Symbol"]==data["Symbol"].shift(-5))&
                  (data["Close"]>data["Close"].shift(-1)))
```
